```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = r"C:\Users\bhavya\OneDrive\Pictures\torm.jpg"  # Replace with your image path
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(original_image)
plt.title("Step 1: Original Image")
plt.axis("off")
plt.show()
```

```python
gray_image = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image, cmap="gray")
plt.title("Step 2: Grayscale Image")
plt.axis("off")
plt.show()
```

```python
inverted_image = cv2.bitwise_not(gray_image)
```

```python
plt.imshow(inverted_image, cmap="gray")
plt.title("Step 3: Inverted Image")
plt.axis("off")
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
```

```python
plt.imshow(blurred_image, cmap="gray")
plt.title("Step 4: Blurred Image")
plt.axis("off")
plt.show()
```

```python
def dodge_blend(front, back):
    return cv2.divide(front, 255 - back, scale=256)
pencil_sketch = dodge_blend(gray_image, blurred_image)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis("off")
```

```python
plt.subplot(1, 2, 2)
plt.title("Pencil Sketch")
plt.imshow(pencil_sketch, cmap="gray")
plt.axis("off")
```

```python
plt.tight_layout()
plt.show()
```
